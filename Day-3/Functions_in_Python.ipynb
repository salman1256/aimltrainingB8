```python
#Variable Length Arguments
def addition(*numbers):
  total=0
  print('\n All Numbers are: ')
  for i in numbers:
    print(i,end=' ')
    total+=i
  return total
```

```python
print('\n Result: ',addition(12,45))
print('\n Result: ',addition(10,30,40))
print('\n Result: ',addition(12,45,89,9,11,23,11,45))
```

## Write a function to take a number from user and print the cube and square of given number

* cube of number means number* number* number
* e.g. cube of 2 is 8 and 3 is 27
* square of number means number*number
* e.g. square of 2 is 4 & 3 is 9

```python
def cube_square(num):
  #print('Cube of',num, 'is', num*num*num)
  print(f'Cube of {num} is {num*num*num}')
  print(f'Square of {num} is {num*num}')
```

```python
given_number=int(input('Enter Number:\t'))
cube_square(given_number)
```

```python
#if C Convert celsius to Fahreneheit
#if F Convert  Fahreneheit to celsius
def convert_temp(temp,unit='C'):
  if unit=='C':
    return (temp*9/5)+32
  elif unit=='F':
    return (temp-32)*5/9
  else:
    return 'Invalid Unit , Please use only C Or F'
```

```python
print('32 ¬∞C is equal to', convert_temp(32),'¬∞F')
```

```python
print('98 ¬∞F is equal to', convert_temp(98,'F'),'¬∞C')
```

```python
print('98 ¬∞C is equal to', convert_temp(98,'C'),'¬∞F')
```
