**List in Python**
Python lists are containers to store a set of values
of any data type.
* List items are ordered, changeable, and allow duplicate values
* Denote the list using square bracket []

```python
emps=['Sam','Ravi','Si','Vi','Amit']
```

```python
print(emps)
```

```python
#indexing start from 0 and goes upto len-1
#len=Number of items in list
print('Number of items in list : ',len(emps))
```

```python
colors=['Red','Yellow','Green','White','Red','Blue']
```

```python
print('Number of items in list : ',len(colors))
```

```python
print('All Items from Emps')
for emp in emps:
    print(emp)
```

```python
print('All Colors')
for color in colors:
  print(color)
```

```python
actors=['Salman Khan','Sharukh Khan','Amir Khan','Amitabh','Ranbir Kappor','Vin Disel','Leo']
```

```python
#indexing start from 0 and goes upto len-1
for i in range(len(actors)):
  print(f"{i}->{actors[i]}")
```

```python
#indexing start from 0 and goes upto len-1
print('Number of actors: ',len(actors))

for(i,actor) in enumerate(actors):
  print(f"{i+1}. {actor}")
```

```python
print('Number of actors: ',len(actors))
for i in range(len(actors)):
  print(f"{i+1}. {actors[i]}")
```
