**Tuple: is an ordered Immutable collection of items**
* we use () to denote tuple

```python
numbers=(100,130,100,230,400,500,100)
```

```python
print('Number of items in tuple are: ',len(numbers))
```

```python
print('All Items in tuple')
for item in numbers:
    print(item,end='\t')
```

```python
for (i,num) in enumerate(numbers):
    print(f'{i}->{num}')
```

```python
# A collection of employees using touple(id,name,salary,designation)
employees=(
    (1,'Sam',5500,'Manager'),
    (2,'Raj',4200,'Developer'),
    (3,'Ravi',3500,'Developer'),
    (4,'Amit',3500,'Developer'),
    (5,'Vini',2500,'Tester'),
    (6,'Deep',5500,'Hr'),
    (7,'Rohan',2600,'Tester'),
)
```

```python
for emp in employees:
    print(f'ID:{emp[0]} \t Name: {emp[1]} \t Salary: {emp[2]} \t Designation: {emp[3]}')
```
