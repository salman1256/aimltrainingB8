**Set**

* Set is an unordered collection of unique items,
 * Its mutable.
* We denote set by {}

```python
first_set={100,200,300,400,500}
print(first_set)
```

```python
second_set={'ac','cooler','fan','washing machine'}
print(second_set)
```

```python
second_set={'ac','cooler','fan','washing machine','ac','cooler','fridge'}
print(second_set)
```

```python
print('Number of items in second set : ',len(second_set))
```

```python
second_set.remove('fan')
print('Set after removing fan: Number of items are: ',len(second_set))
print(second_set)
```

```python
second_set.add('mobile')
print('Set after adding mobile: Number of items are: ',len(second_set))
print(second_set)
```

```python
second_set.clear()
print('Set after clearing all items: Number of items are: ',len(second_set))
print(second_set)
```

```python
second_set.add('Laptop')
print('Set after adding Laptop: Number of items are: ',len(second_set))
print(second_set)
```

```python
set_one={1,2,3,4,5}
set_two={5,6,4}
```

```python
print('Number of items in set_one',len(set_one))
print('Number of items in set_two',len(set_two))
common=set_one.intersection(set_two)
print('Common items in both sets: ',common)
```

```python
all_items=set_one.union(set_two)
print('All items in both sets: ',all_items)
#
```

```python
diff_items=set_one.difference(set_two)
print('Items in set_one but not in set_two: ',diff_items)
#
```

```python
course_one={'Python','Power BI','Excel','SQL','Time Management','Soft Skill'}
```

```python
course_two={'Python','Excel','SQL','Soft Skill','Azure','GitHub','Tensorflow'}
```

```python
print('Number of items in course_one: ',len(course_one))
for item in course_one:
  print(item)
```

```python
print('Number of items in course_two: ',len(course_two))
for item in course_two:
  print(item)
```

```python
common_topics=course_one.intersection(course_two)
print('Number ofCommon topics in both courses: ',len(common_topics))
print('Common topics in both courses: ',common_topics)
```

```python
both_courses=course_one.union(course_two)
print('Number of topics in both courses: ',len(both_courses))
print('Topics in both courses: ',both_courses)
#
```

```python
course_one_diff=course_one.difference(course_two)
print('Topics in course_one but not in course_two: ',len(course_one_diff))
print('Topics in course_one but not in course_two: ',course_one_diff)
#
```

```python
course_two_diff=course_two.difference(course_one)
print('Topics in course_two but not in course_one: ',len(course_two_diff))
print('Topics in course_two but not in course_one: ',course_two_diff)
#
```
