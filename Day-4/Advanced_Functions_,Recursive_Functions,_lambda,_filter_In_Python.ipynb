```python
#Recursive functions: If the function is callling to itself ,It's known as recursive functions
```

```python

def factorial(num):
  if(num==1):
    return 1
  else:
    return num*factorial(num-1)
```

```python
factorial(1)
```

```python
factorial(3)
```

```python
factorial(5)
```

```python
factorial(6)
#6*5*4*3*2*1
```

```python
#Recursive function Example -2
def reverse_string(text):
  if len(text)==0:
    return ''
  else:
    return (reverse_string(text[1:])+text[0])
```

```python
print('Reverse of "am" is: \t',reverse_string('am'))
```

```python
print('Reverse of "pria" is: \t',reverse_string('pria'))
```

```python
user_text=input('Enter text to reverse')
print('Reverse of',user_text,'is:',reverse_string(user_text))
```
