```python
def calc_price(price, discount=0.05):

  return (price - (price * discount))

```

```python
def calc_price(price, discount=0.1):
  if(price>=50):
    discount=0.3
  elif(price>=25):
    discount=0.2

  price_after_discount=price-(price*discount)
  return round(price_after_discount,2)
```

```python
pizza_price=50.25
doughnut_price=5
small_cake=33
cake_price=120.50
print(f'Pizza price {pizza_price} after discount Price:{calc_price(pizza_price)}')
print(f'Doughnut price {doughnut_price} after discount Price:{calc_price(doughnut_price)}')
print(f'Cake price {cake_price} after discount Price:{calc_price(cake_price)}')
print(f'Small Cake price {small_cake} after discount Price:{calc_price(small_cake)}')
```
