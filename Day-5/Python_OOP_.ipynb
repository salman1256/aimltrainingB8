```python
#Create a class for students , class have following attributes id, name, course, school
class Student:
  def __init__(self,id,name,course,school):
    self.id=id
    self.name=name
    self.course=course
    self.school=school
  def __str__(self):
    return f'Student Id: {self.id}\t Student Name: {self.name} \t Student Course: {self.course} \t Student School: {self.school}'
```

```python
#Create 5 instances(objects) of class
stud1=Student(1,'Sam','Fifth','DAV')
stud2=Student(2,'Tom','Sixth','DAV')
stud3=Student(3,'Bob','Seventh','RPS')
stud4=Student(4,'Tim','Eighth','DPS')
stud5=Student(5,'Jack','Ninth','DAV')
```

```python
#Put in list of students and display it.
students=[stud1,stud2,stud3,stud4,stud5]
for s in students:
  print(s)
```

```python
class BankAccount:
  def __init__(self,account_number,account_holder,balance):
    self.account_number=account_number
    self.account_holder=account_holder
    self.balance=balance
    print('Account Created')

  def display_balance(self):
    print('Account Balance:',self.balance)

  def deposit(self,amount):
    self.balance+=amount
    print('Amount Deposited')
    print('Current Balance:',self.balance)

  def withdraw(self,amount):
    if self.balance>=amount:
      self.balance-=amount
      print('Amount Withdrawn')
      print('Current Balance:',self.balance)
    else:
      print('Insufficient Balance')

  def __str__(self):
    return f'Account Number: {self.account_number} \t Account Holder: {self.account_holder} \t Balance: {self.balance}'
```

```python
account1=BankAccount('AC-001','Sam',12000)
```

```python
account2=BankAccount('AC-002','Tom',15000)
```

```python
account3=BankAccount('AC-11300','Arsh',20000)
```

```python
account1.display_balance()
account1.deposit(2000)
account1.withdraw(5000)
account1.display_balance()
print(account1)
```
