**Exercise: **
Write a number game to find out hidden number between (1 to 10).


1.   Ask the user to guess the hidden number
*. Note: Hidden Number is 7
2.   if he entered number leass than 7 show too Low
3.  If he entered number more than 7 show too High
4. if he entered correct number show Correct Answer
 You have won and end the loop
5. Only 3 attempts are allowed
6. If he have not guess within 3 attempts show You have lost all your attempts






```python
Hidden_number=7
allowed_attempts=3

while(allowed_attempts!=0):
  user_guess=int(input('Enter your guess:\t'))
  if(user_guess==Hidden_number):
    print('You have won')
    break

  else:


    allowed_attempts-=1

    if(allowed_attempts==0):
      print('You have lost all your attempts')
    else:
      if(user_guess>Hidden_number):
        print('Too High')
      else:
        print('Too Low')
```
